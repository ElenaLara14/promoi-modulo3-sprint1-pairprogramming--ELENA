```python
# 4 EJERCICIO

def classify_infant_mortality(mortality_rate): 
    if mortality_rate < 10: 
        return 'Baja' 
    elif mortality_rate < 40: 
        return 'Moderada' 
    else: 
        return 'Alta' 
# Aplicar la clasificación 
df['infant_mortality_category'] = df['infant_mortality'].apply(classify_infant_mortality) 
# Crear la lista de países con sus categorías 
result = (df[['country', 'infant_mortality_category']])
```

```python
df1 = pd.DataFrame(result)
df1
```

```python
# 5 EJERCICIO

columnas = ["fertility_rate", "life_expectancy"]
df_col = df[columnas]
df_correlaciones = df_col.corr(method= "spearman")
df_correlaciones
```

```python
sns.scatterplot(x="fertility_rate", y="life_expectancy", data=df_correlaciones)
```

```python
sns.heatmap( df_correlaciones,annot=True,fmt=".2f")
```
