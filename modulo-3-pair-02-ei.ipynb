- Contexto: Trabajas en una organización de salud global y deseas estimar el intervalo de confianza al 95% para la esperanza de vida promedio de cada continente.
    Nota: Cuidado porque puede que os salgan algunos nulos, genera un código para que no salgan nulos en los intervalos de confianza.

- Intervalo de Confianza para el PIB

- Contexto: Eres una economista y quieres calcular el intervalo de confianza al 90% para el Producto Interno Bruto (PIB) de los continentes.

```python
df.isna().sum()
```

```python
media_ev = df["life_expectancy"].mean()
error = stats.sem(df["life_expectancy"], nan_policy="omit")
nivel_confianza = 0.95
g_libertad = len(df["life_expectancy"])-1
v_crit = stats.t.ppf((1+nivel_confianza)/2, df = g_libertad)

l_superior = media_ev + v_crit * error
l_inferior = media_ev - v_crit * error

print (l_inferior, "-->", l_superior)
```

```python
def calcular_intervalo_confianza(grupo):

    media_ev = grupo["life_expectancy"].mean()
    error = stats.sem(grupo["life_expectancy"], nan_policy="omit")
    nivel_confianza = 0.95
    g_libertad = len(grupo["life_expectancy"]) - 1
    v_crit = stats.t.ppf((1 + nivel_confianza) / 2, df=g_libertad)
    
    l_superior = media_ev + v_crit * error
    l_inferior = media_ev - v_crit * error
    
    return pd.Series([l_inferior, l_superior], index=["l_inferior", "l_superior"])

# Agrupar por continente y aplicar la función para calcular intervalos de confianza
df.groupby('continent').apply(calcular_intervalo_confianza).reset_index()
```
